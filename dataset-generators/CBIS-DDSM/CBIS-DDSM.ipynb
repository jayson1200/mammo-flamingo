```python
laterality_syms_in_data = np.unique(df_calc_test["left or right breast"].to_numpy())
laterality_ops = list(map(str.lower, laterality_syms_in_data))

lat_choices_dict = dict(zip(laterality_syms_in_data, laterality_ops))
lat_choices_dict
```

```python
density_choices_dict = {
    0: "incomplete",
    1: "almost entirely fat",
    2: "containing scattered areas of fibroglandular density",
    3: "heterogenously dense",
    4: "extremely dense"
}

density_choices_dict
```

```python
calc_type_syms_in_data = np.unique(df_calc_test["calc type"].to_numpy())
calc_type_ops = list(map(str.lower, calc_type_syms_in_data))

calc_type_choices_dict = dict(zip(calc_type_syms_in_data, calc_type_ops))

calc_type_choices_dict
```

```python
calc_dist_syms_in_data = np.unique(df_calc_test["calc distribution"].to_numpy())
calc_dist_ops = list(map(str.lower, calc_dist_syms_in_data))

calc_dist_choices_dict = dict(zip(calc_dist_syms_in_data, calc_dist_ops))

calc_dist_choices_dict
```

```python
assessment_syms_in_data = np.unique(df_calc_test["assessment"].to_numpy())

assessment_choices_dict = {0:"zero", 1:"one", 2:"two", 3:"three", 4:"four", 5:"five", 6:"six" }

assessment_choices_dict
```

```python
pathology_syms_in_data = np.unique(df_calc_test["pathology"].to_numpy())
pathology_ops = list(map(str.lower, pathology_syms_in_data))

pathology_choices_dict = dict(zip(pathology_syms_in_data, pathology_ops))
pathology_ops
```

```python
print(np.unique(df_calc_test["subtlety"].to_numpy()))

subtlety_choices_dict = {1:"one", 2:"two", 3:"three", 4:"four", 5:"five"}
```

```python
calc_test_qs = []
for idx, row in df_calc_test.iterrows():
    curr_pat_qs = {}
    
    curr_pat_qs["laterality"] = {
        "Q": "Which breast is this left or right?",
        "A": row["left or right breast"]
    }

    curr_pat_qs["density"] = {
        "Q": "What is breast density show here?",
        "A": row["breast density"]
    }

    curr_pat_qs["image view"] = {
        "Q": "What view was used when taking the mammogram?",
        "A": row["image view"]
    }

    curr_pat_qs["calc type"] = {
        "Q": "What is the calcification type show here?",
        "A": row["calc type"]
    }
    
    curr_pat_qs["calc distribution"] = {
        "Q": "What is the calcification distribution shown here?",
        "A": row["calc distribution"]
    }
    
    curr_pat_qs["subtlety"] = {
        "Q": "On a scale of 1 to 5, describe the subtlety of the abnormality?",
        "A": row["subtlety"]
    }

    curr_pat_qs["assesssment"] = {
        "Q": "What is you BI-RADS assessment rating?",
        "A": row["assessment"]
    }

    curr_pat_qs["image folder"] = row["image file path"].split("/")[0]

    calc_test_qs.append(curr_pat_qs)


choices = {
    "laterality": lat_choices_dict,
    "density": density_choices_dict,
    "image view": image_view_choices_dict,
    "calc type": calc_type_choices_dict,
    "calc distribution": calc_dist_choices_dict,
    "subtlety": subtlety_choices_dict,
    "assessment": assessment_choices_dict
}
res = {"Qs": calc_test_qs, 'choices':choices}
```

```python
with open("../datasets/calc_test.json", "w") as file:
    json.dump(res, file, indent=4)
```
